```python
import pandas as pd
import numpy as np
import requests
import yfinance as yf
import time
```

```python
from google.colab import drive
import os

drive.mount('/content/drive')

project_path = '/content/drive/MyDrive/stock-vs-crypto-risk-return-analysis/data'
os.makedirs(project_path, exist_ok=True)
```

```python
tickers = [
    "AAPL", "MSFT", "TSLA", 'AMZN', 'GOOGL', # Global MNCs
    'RELIANCE.NS', 'INFY.NS', 'TCS.NS' # Indian MNCs
    ]
stock_data = {}

for ticker in tickers:
  df = yf.download(ticker, period="5y", interval="1d")
  df.reset_index(inplace=True)
  stock_data[ticker] = df[["Date", "Close"]]
```

```python
# Merge into one CSV file
stocks_merge = pd.concat(
    [df.assign(Ticker = ticker) for ticker, df in stock_data.items()]
)

stocks_merge.to_csv(f'{project_path}/raw_stocks.csv', index=False)
```

```python
stocks_wide = stocks_merge.pivot_table(
    index='Date',
    columns='Ticker',
    values='Close',
    aggfunc='mean'  # or 'first', 'max', 'min' depending on your use case
)
stocks_wide.to_csv(f'{project_path}/stocks_wide.csv')
stocks_wide.head()
```
